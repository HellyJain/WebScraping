```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
def splt_list(l, n):
    #Split long list
    n = max(1, n)
    return list(l[i:i+n] for i in range(0, len(l), n))
```

Scrapping the latest population of the countries in the world from worldometer

```python
def scrap_world_pop():
    URL = 'https://www.worldometers.info/geography/alphabetical-list-of-countries/'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    header = [h.text for h in soup.find_all("th")]
    countries = [result.text for result in soup.find_all("td")]
    countries = splt_list(countries, 5)
    return pd.DataFrame(countries, columns = header)
df_countries = scrap_world_pop()
df_countries.tail(15)
```

Scrapping the COVID cases for continents and countries from worldometer

```python
def scrap_covid_world():
    URL = 'https://www.worldometers.info//coronavirus/'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    header = [h.text for h in soup.find_all("th")]
    countries = [result.text for result in soup.find_all("td")]
    header = header[:13]
    countries = splt_list(countries, 13)
    return pd.DataFrame(countries, columns = header)
    
df_covid_world = scrap_covid_world()
df_covid_world.loc[8:23]
```
